## **Functions**

### Scalar function

Defining the functions

```sql
CREATE OR REPLACE FUNCTION
catalog_sql.sales.discounted_price (p_price DECIMAL(10,2))
RETURNS DECIMAL(10,2)
RETURN p_price * 0.9
```

```sql
SELECT price_per_unit,
       catalog_sql.sales.discounted_price(price_per_unit) AS discounted_price
FROM catalog_sql.sales.order
```

### Table function

```sql
CREATE OR REPLACE FUNCTION catalog_sql.sales.discount_by_category(p_category STRING)
RETURNS TABLE
LANGUAGE SQL
RETURN
(SELECT *
FROM catalog_sql.sales.order
WHERE product_category = p_category)
```

```sql
SELECT * FROM catalog_sql.sales.discount_by_category('Home Decor')
```
