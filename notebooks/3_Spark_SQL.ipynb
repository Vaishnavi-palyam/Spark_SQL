## **Merge/Upsert**

UPSERT = Update + Insert

Update the existing rows

Insert the new rows

Read the table for source

```sql
%python
df = spark.read.table('catalog_sql.sales.order')
```

Convert the df into view to perform SQL merge

```sql
%python
df.createOrReplaceTempView('order_source')
```

Merge

```sql
MERGE INTO catalog_sql.sales.order_ext trg
USING order_source as src
on trg.order_id = src.order_id
WHEN MATCHED THEN
UPDATE SET *
WHEN NOT MATCHED THEN
INSERT *
```
