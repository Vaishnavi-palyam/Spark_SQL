# **Row level security**

### Mapping table

```sql
CREATE TABLE catalog_sql.sales.map_table
(
  payment_method STRING,
  email STRING
)
```

```sql
SELECT * FROM catalog_sql.sales.`order`
```

```sql
INSERT INTO catalog_sql.sales.map_table
VALUES
('Credit Card','palyamvaishnavi@gmail.com'),
('Debit Card','palyamvaishnavi@gmail.com'),
('PayPal','newuser@gmail.com'),
('UPI','newuser@gmail.com')
```

```sql
SELECT * FROM catalog_sql.sales.map_table
WHERE email = current_user()
```

### Converting mapping table into a boolean function

```sql
SELECT exists
(
SELECT 1 FROM catalog_sql.sales.map_table
where email = current_user()
and payment_method = 'Credit Card'
)
```

Convert into a boolean function

```sql
CREATE OR REPLACE FUNCTION catalog_sql.sales.rls(p_column STRING)
RETURNS BOOLEAN
LANGUAGE SQL
return (exists
            (
              SELECT 1 FROM catalog_sql.sales.map_table
              where email = current_user()
              and payment_method = p_column
              ))
```

Adding the RLS to the table

```sql
ALTER TABLE catalog_sql.sales.`order`
SET ROW FILTER catalog_sql.sales.rls
ON (payment_method)
```

```sql
SELECT * FROM catalog_sql.sales.`order`
```
